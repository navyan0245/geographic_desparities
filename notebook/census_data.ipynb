```python
import pandas as pd
```

```python
census_2021=pd.read_csv('../data/2021_census.csv')
```

```python
census_2021.head(7)
```

```python
rowCntr = 0
row1 = None
df = pd.DataFrame(columns=['Year', 'County', 'Population_Type', 'Population', 'Percent', 'Income'])

for i, item in census_2021.iterrows():
    if rowCntr % 7 == 0:
        row1 = item
    if rowCntr % 7 == 2:
        row3 = item
    if rowCntr % 7 == 4:
        row5 = item
    if rowCntr % 7 == 6:
        row7 = item

        df.loc[len(df)] = {
            'Year': row1.Year, 'County': row1.County, 'Population_Type': 'Total_population', 
            'Population': row3.Total_population, 
            'Percent' : '100%',
            'Income': row7.Total_population
        }

        df.loc[len(df)] = {
            'Year': row1.Year, 'County': row1.County, 'Population_Type': 'White', 'Population': row3.White, 
            'Percent' : row5.White,
            'Income': row7.White
        }

        df.loc[len(df)] = {
            'Year': row1.Year, 'County': row1.County, 'Population_Type': 'Black', 'Population': row3.Black, 
            'Percent' : row5.Black,
            'Income': row7.Black
        }

        df.loc[len(df)] = {
            'Year': row1.Year, 'County': row1.County, 'Population_Type': 'American_Indian_and_Alaska_Native', 
            'Population': row3.American_Indian_and_Alaska_Native, 
            'Percent' : row5.American_Indian_and_Alaska_Native,
            'Income': row7.American_Indian_and_Alaska_Native
        }

        df.loc[len(df)] = {
            'Year': row1.Year, 'County': row1.County, 'Population_Type': 'Asian', 
            'Population': row3.Asian, 
            'Percent' : row5.Asian,
            'Income': row7.Asian
        }

        df.loc[len(df)] = {
            'Year': row1.Year, 'County': row1.County, 'Population_Type': 'Native_Hawaiian_and_Other_Pacific_Islander', 
            'Population': row3.Native_Hawaiian_and_Other_Pacific_Islander, 
            'Percent' : row5.Native_Hawaiian_and_Other_Pacific_Islander,
            'Income': row7.Native_Hawaiian_and_Other_Pacific_Islander
        }

        df.loc[len(df)] = {
            'Year': row1.Year, 'County': row1.County, 'Population_Type': 'other_race', 
            'Population': row3.other_race, 
            'Percent' : row5.other_race,
            'Income': row7.other_race
        }

        df.loc[len(df)] = {
            'Year': row1.Year, 'County': row1.County, 'Population_Type': 'Hispanic_or_Latino_origin', 
            'Population': row3.Hispanic_or_Latino_origin, 
            'Percent' : row5.Hispanic_or_Latino_origin,
            'Income': row7.Hispanic_or_Latino_origin
        }
        
    rowCntr += 1

```

```python
print(census_2021.columns.tolist())
```

```python
df.to_csv('../data/2021_census_out.csv', index=False)
```
